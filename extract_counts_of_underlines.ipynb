```python
import pandas as pd
```

```python
p = 'underlined.csv'
```

```python
df = pd.read_csv(p)
```

```python
df
```

```python
df['u_lower'] = df['underline'].str.lower()
```

```python
l = """same
the same
another
Idem
id
vid
one
ivory id
d
lesser one
smaller one
larger one
ones
one finer
One round
One larger
One hemisphericall
One agate & red Jasper
ib
:Ib
shoe Id
it
The same wood"""
```

```python
l = l.split('\n')
l = [x.lower() for x in l]
```

```python
l
```

```python
cats = ['Miscellanies.', 'Antiquities.', 'Impressions of Seals &c.',
       'Pictures &c.', 'Mathematicall instruments &c', 'Agate handles &c.',
       'Agate cups botles spoons &c.']
```

```python
df
```

```python
df_res = pd.DataFrame({'Category': [], 'Word': [], 'Count': []})
for c in cats:
    df_sub = df[df.label == c]
#     print(df_sub)
    print(c)
    for word in l:
        print(word)
        word = "'" + word + "'"
        print(word)
        print(df_sub['u_lower'].str.contains(word).sum())
        count = df_sub['u_lower'].str.contains(word).sum()
        df_res = df_res.append({'Category': c, 'Word': word, 'Count': count}, ignore_index=True)
    print('***')
# df_res
```

```python
df_res.to_csv('underlined_counts_by_label2.csv')
```
