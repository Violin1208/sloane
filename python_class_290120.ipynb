```python
#works with a perfec match
import numpy as np

n=[5,6,9,8,7,4,3,9]
m=[4,3,9,8,5,7,6]

for i in range(1, len(m)):
    #for m from the begging. n from the end
    n_temp=n[-i:]
    m_temp=m[i]
    #to find a perfect match
    if(n_temp == m_temp):
        print(i)
        break
```

```python
#thinga are not perfect matches. So using this way you can overcome this issue.
import numpy as np

n=[5,6,9,8,7,4,3,9]
m=[4,3,9,8,5,7,6]

n=np.array(n)
m=np.array(m)

for i in range(1, len(m)):
    #for m from the begging. n from the end
    n_temp=n[-i:]
    m_temp=m[i]
    #to find a perfect match
    if(np.mean(n_temp == m_temp) > 0.90):
        print(i)
```

```python
#working out where a match occurs using numpy so that it works even there is no 100% match. This method can also be used to match photos.

import numpy as np

d=[1,2,1,3,2,2,3,4,1,2,3,4,4,3,2]
g=[1,2,3,4]

d=np.array(d)
g=np.array(g)
```

```python
for i in range(len(d)-len(g)+1):
    t=d[i:i+len(g)]
    #print(t)
    
    k=(t==g)
    if (np.sum(k)==len(g)):
        print(i)
```

```python
#Likelyhood. 
likelyhood = []
for i in range(len(d)-len(g)+1):
    t=d[i:i+len(g)]
    
    k=(t==g)
    likelyhood.append(100*np.mean(k))
```

```python
#to see edge of an image

import numpy as np
import matplotlib.pyplot as plt

#when matplotlib used, it's compatible with numpy. Pandas not compatible with numpy. So you will have to convert to np.array
image=plt.imread('/Users/deborahleem/sample-image-edge-detection-200129.png')
#average
image=(image[:,:,0] + image[:,:,1] + image[:,:,2])/3

plt.imshow(image)

edge=np.zeros(image.shape)

image_size=image.shape
for count_row in range(image_size[0]):
    for count_column in range(image_size[1]-1):
        if(image[count_row, count_column]-image[count_row, count_column+1]> 5/255):
            edge[count_row, count_column]=1
plt.imshow(edge)
```
